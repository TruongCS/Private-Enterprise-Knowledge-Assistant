 # 2.Load the Markdown report

```python
from pathlib import Path
```

```python
# Change this if your Markdown file has a different name.
REPORT_MD = Path('uber_report.md')

if not REPORT_MD.exists():
    raise FileNotFoundError(f'Markdown report not found: {REPORT_MD}')

text = REPORT_MD.read_text(encoding='utf-8')
print(f'Loaded {REPORT_MD.name}')
print(f'Characters: {len(text):,}')
print(text[:500])
```

# 3.Split the Report into Search Chunks

```python
import re
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter

CHUNK_SIZE = 500
CHUNK_OVERLAP = 50

# Remove Markdown tables from vector chunks because tables go into SQLite separately.
text_without_tables = re.sub(r'((?:\|.+\n)+)', '', text)

splitter = RecursiveCharacterTextSplitter(
    chunk_size=CHUNK_SIZE,
    chunk_overlap=CHUNK_OVERLAP,
)

chunks = splitter.split_text(text_without_tables)
docs = [Document(page_content=chunk, metadata={'source': REPORT_MD.name}) for chunk in chunks]

print(f'Created {len(docs):,} text chunks')
print(docs[0].page_content[:500])
```

# 4. Build and Save the FAISS Vectorestore


```python
VECTORSTORE_DIR = "/content/drive/MyDrive/FiniancialAgent/vectorstore"
```

```python
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings

EMBEDDING_MODEL = 'all-MiniLM-L6-v2'

embedding = HuggingFaceEmbeddings(model_name=EMBEDDING_MODEL)
vectorstore = FAISS.from_documents(docs, embedding)
vectorstore.save_local(str(VECTORSTORE_DIR))

print(f'Saved vectorstore to: {VECTORSTORE_DIR}')
```

# 4.Load the Vectorstore and Create a Hybrid Retriever

The hybrid retriever combines:

*   FAISS semantic search for meaning-based matches.
*   BM25 keyword search for exact terms.


```python
from langchain_community.retrievers import BM25Retriever
from langchain_classic.retrievers import EnsembleRetriever

vectorstore = FAISS.load_local(
    str(VECTORSTORE_DIR),
    embedding,
    allow_dangerous_deserialization=True,
)

dense_retriever = vectorstore.as_retriever(search_kwargs={'k': 3})
sparse_retriever = BM25Retriever.from_documents(docs)
sparse_retriever.k = 3

hybrid_retriever = EnsembleRetriever(
    retrievers=[dense_retriever, sparse_retriever],
    weights=[0.7, 0.3],
)

print('Hybrid retriever ready.')
```

# 5.Test Retrieval Before Using the LLM

```python
query = "What is Uber's revenue?"
results = hybrid_retriever.invoke(query)

for i, doc in enumerate(results, start=1):
    print(f'\n--- Document {i} ---')
    print(doc.page_content[:1000])
```

# 7.Extract Markdown Tables into SQLite

```python
from pathlib import Path
import sqlite3
import pandas as pd
```

```python
DB_PATH = Path("/content/drive/MyDrive/FiniancialAgent/uber_financials.db")

# Check the folder exists
if not DB_PATH.parent.exists():
    raise FileNotFoundError(f"Folder does not exist: {DB_PATH.parent}")

# This creates the .db file if it does not already exist
con = sqlite3.connect(DB_PATH)
con.close()

print(f"SQLite database ready: {DB_PATH}")
```

```python
def safe_name(value: str, fallback: str) -> str:
    """Convert a heading or column into a SQLite-friendly name."""
    name = re.sub(r'\W+', '_', value).lower().strip('_')
    return name or fallback


def reset_sqlite(db_path: Path) -> None:
    """Drop old tables so the database matches the current Markdown report."""
    con = sqlite3.connect(db_path)
    tables = con.execute("SELECT name FROM sqlite_master WHERE type='table'").fetchall()
    for (table_name,) in tables:
        escaped_name = table_name.replace('"', '""')
        con.execute(f'DROP TABLE IF EXISTS "{escaped_name}"')
    con.commit()
    con.close()


def extract_tables_to_sqlite(markdown: str, db_path: Path, source_name: str) -> None:
    """Find Markdown tables and save them as SQLite tables."""
    con = sqlite3.connect(db_path)
    pattern = re.compile(r'(?:#{1,3}\s*(.+?)\n)?((?:\|.+\n)+)', re.MULTILINE)
    source_prefix = safe_name(Path(source_name).stem, 'source')
    saved, skipped = 0, 0

    for i, match in enumerate(pattern.finditer(markdown)):
        heading = (match.group(1) or f'table_{i}').strip()
        table_md = match.group(2)

        lines = [line.strip() for line in table_md.strip().split('\n') if line.strip()]
        lines = [line for line in lines if not re.match(r'^\|[\s\-|]+\|$', line)]
        rows = [[cell.strip() for cell in line.strip('|').split('|')] for line in lines]

        if len(rows) < 2:
            skipped += 1
            continue

        headers = [safe_name(col, f'col_{j}') for j, col in enumerate(rows[0])]
        seen = {}
        deduped_headers = []
        for header in headers:
            seen[header] = seen.get(header, 0) + 1
            deduped_headers.append(header if seen[header] == 1 else f'{header}_{seen[header]}')

        df = pd.DataFrame(rows[1:], columns=deduped_headers)
        table_slug = safe_name(heading[:50], f'table_{i}')
        table_name = f'{source_prefix}_{table_slug}'
        df.to_sql(table_name, con, if_exists='replace', index=False)
        saved += 1

    con.close()
    print(f'Saved {saved} tables to {db_path}; skipped {skipped} small/invalid tables.')


reset_sqlite(DB_PATH)
extract_tables_to_sqlite(text, DB_PATH, REPORT_MD.name)
```

# 8.Inspect the SQLite Tables

Run this to see which tables were extracted and preview their columns

```python
def list_tables(db_path: Path) -> list[str]:
    con = sqlite3.connect(db_path)
    tables = pd.read_sql_query(
        "SELECT name FROM sqlite_master WHERE type='table' ORDER BY name",
        con,
    )
    con.close()
    return tables['name'].tolist()


def preview_tables(db_path: Path, limit: int = 3) -> None:
    con = sqlite3.connect(db_path)
    for table_name in list_tables(db_path):
        print(f'\n=== {table_name} ===')
        df = pd.read_sql_query(f'SELECT * FROM "{table_name}" LIMIT {limit}', con)
        print(df.to_string(index=False))
    con.close()


tables = list_tables(DB_PATH)
print(f'Table count: {len(tables)}')

# see detailed preview of every table
preview_tables(DB_PATH)
```

# 9.Create Tools for the Financial Agent

These tools let the agent retrieve narrative context, search table names, query SQLite, and calculate derived numbers

```python
from langchain_core.tools import tool


@tool
def retrieve_financial_context(query: str) -> str:
    """Search the annual report for narrative information such as strategy, risks, and business descriptions."""
    retrieved = hybrid_retriever.invoke(query)
    if not retrieved:
        return 'No relevant content found.'
    return '\n\n---\n\n'.join(doc.page_content for doc in retrieved)


@tool
def list_available_tables(_: str = '') -> str:
    """List all financial tables available in SQLite."""
    return '\n'.join(list_tables(DB_PATH))


@tool
def search_tables_for_keyword(keyword: str) -> str:
    """Search SQLite tables for a keyword in column names or sample rows."""
    con = sqlite3.connect(DB_PATH)
    matches = []

    for table_name in list_tables(DB_PATH):
        try:
            df = pd.read_sql_query(f'SELECT * FROM "{table_name}" LIMIT 5', con)
            searchable_text = ' '.join(
                df.columns.astype(str).tolist() + df.astype(str).values.flatten().tolist()
            ).lower()
            if keyword.lower() in searchable_text:
                matches.append(f'{table_name}: columns = {df.columns.tolist()}')
        except Exception:
            continue

    con.close()
    return '\n'.join(matches) if matches else f"No tables found containing '{keyword}'."


@tool
def query_financial_table(sql: str) -> str:
    """Run a SELECT query against the financial SQLite tables."""
    if re.search(r'\b(insert|update|delete|drop|alter|create)\b', sql, re.IGNORECASE):
        return 'Only SELECT queries are allowed.'

    con = sqlite3.connect(DB_PATH)
    try:
        df = pd.read_sql_query(sql, con)
        return 'No rows returned.' if df.empty else df.to_markdown(index=False)
    except Exception as exc:
        return f'SQL error: {exc}'
    finally:
        con.close()


@tool
def calculate(expression: str) -> str:
    """Evaluate arithmetic for growth rates, ratios, and totals."""
    if not re.match(r'^[\d\s\.\+\-\*\/\(\)\%]+$', expression):
        return 'Invalid expression. Only numbers and math operators are allowed.'
    try:
        return str(round(eval(expression, {'__builtins__': {}}), 4))
    except Exception as exc:
        return f'Calculation error: {exc}'


tools = [
    retrieve_financial_context,
    list_available_tables,
    search_tables_for_keyword,
    query_financial_table,
    calculate,
]

print(f'Created {len(tools)} tools.')
```

# 10.Build the Financial Agent

The system prompt tells the agent when to use SQLite versus vector retrieval

```python
from langchain_classic.agents import create_tool_calling_agent, AgentExecutor
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage

system_prompt = """
You are a precise financial analyst assistant specialising in Uber's annual report.

Use the right tool for each job:
- Use list_available_tables to see all available SQLite tables.
- Use search_tables_for_keyword FIRST for numerical questions when you do not know the right table.
- Use query_financial_table for exact financial values from tables. Always inspect columns first with SELECT * FROM table LIMIT 3.
- Use retrieve_financial_context for narrative questions: strategy, risks, business model, or qualitative descriptions.
- Use calculate only to derive a new number such as a growth rate or ratio.

For numerical questions:
1. Search tables by keyword.
2. Inspect the likely table schema.
3. Query the exact row/columns.
4. Calculate only if the requested answer is not already shown.
5. Report exact figures from the data. Never invent or approximate numbers.
""".strip()

agent_prompt = ChatPromptTemplate.from_messages([
    ('system', system_prompt),
    MessagesPlaceholder(variable_name='chat_history'),
    ('human', '{input}'),
    MessagesPlaceholder(variable_name='agent_scratchpad'),
])

agent = create_tool_calling_agent(llm=llm, tools=tools, prompt=agent_prompt)
agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=True,
    max_iterations=15,
    handle_parsing_errors=True,
)

print('Financial agent ready.')
```

# 11.Ask questions

Use ask(...) for normal questions. Use reset_memory() when you want a fresh conversation

```python
chat_history = []


def ask(question: str) -> str:
    """Ask the financial agent a question and keep conversation memory."""
    response = agent_executor.invoke({
        'input': question,
        'chat_history': chat_history,
    })
    answer = response['output']
    chat_history.append(HumanMessage(content=question))
    chat_history.append(AIMessage(content=answer))
    return answer


def reset_memory() -> None:
    """Clear conversation history."""
    global chat_history
    chat_history = []
    print('Memory cleared.')
```
